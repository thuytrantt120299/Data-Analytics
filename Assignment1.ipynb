### 1.2 Merge

```python
# Merge data1 and data2 based on student ID
merged_data = pd.merge(data1, data2, on="ID")
merged_data
```

```python
# Concatenate merged_data and data3 based on student ID
final_data = pd.concat([merged_data, data3], ignore_index=True, sort=False)
final_data
```

### 1.3 Clean

```python
#Check for missing values
final_data.isnull().sum()
```

```python
# Calculate the percentage of missing values in each column
missing_perc = final_data.isnull().sum() / len(final_data)

# Find columns with more than 50% missing values
cols_to_drop = missing_perc[missing_perc > 0.5].index

# Find rows with more than 50% missing values
rows_to_drop = final_data.loc[final_data.isnull().sum(axis=1) / len(final_data.columns) > 0.5].index

# Drop the columns and rows with more than 50% missing values
final_data.drop(cols_to_drop, axis=1, inplace=True)
final_data.drop(rows_to_drop, axis=0, inplace=True)
```

```python
# replace missing values in numerical columns with mean
for col in final_data.select_dtypes(include=[np.number]):
    if final_data[col].isnull().sum() > 0:
        final_data[col].fillna(final_data[col].mean(), inplace=True)

# replace missing values in categorical columns with mode
for col in final_data.select_dtypes(include=['object']):
    if final_data[col].isnull().sum() > 0:
        final_data[col].fillna(final_data[col].mode()[0], inplace=True)
        
# check missing values again to ensure all missing values are replaced
print(final_data.isnull().sum())
```

```python
#Check for duplicates in the dataset based on the ID column
duplicates = merged_data.duplicated(subset=['ID'])
print(duplicates.sum())
```

```python
#Check for impossible values:
#Remove rows where age is greater than 100 
final_data = final_data[final_data["Age at enrollment"] <= 100]
final_data
```

```python
#Remove extra whitespaces:
final_data = final_data.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```
